```python
from netCDF4 import Dataset
# wrfout.nc verisine erişim için farklı yollar bulunabilir. exp : xarray
ncfile = Dataset('data/wrfout_d01_2020-12-07_18.nc')
# ncfile..... özellikleri değişken sayısını arttırmak ve otonom çalışma için önemli
```

```python
from wrf import get_basemap
bm = get_basemap(wrfin=ncfile)
```
